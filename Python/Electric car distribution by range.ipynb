```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv(r"Electric_range_among_categories.csv")

colors = {"Battery Electric Vehicle (BEV)": "blue", "Plug-in Hybrid Electric Vehicle (PHEV)": "gold"}

mean_bev_range = data[data['Electric_Vehicle_Type'] == "Battery Electric Vehicle (BEV)"]['Electric_Range'].mean()
mean_phev_range = data[data['Electric_Vehicle_Type'] == "Plug-in Hybrid Electric Vehicle (PHEV)"]['Electric_Range'].mean()

plt.figure(figsize=(10, 6))

ax = sns.violinplot(x='Electric_Vehicle_Type', y='Electric_Range', data=data, palette=colors, inner=None)

plt.xlabel('')
plt.ylabel(r'Electric Range (Miles)')

ax.plot(0, mean_bev_range, marker='o', markersize=10, color='black', label='Mean (BEV)')
ax.plot(1, mean_phev_range, marker='o', markersize=10, color='black', label='Mean (PHEV)')

plt.show()
```
